Hough Circle Transform 
======================

Goal
----

In this chapter,
    -   We will learn to use Hough Transform to find circles in an image.
    -   We will see these functions: **cv.HoughCircles()**

Theory
------

A circle is represented mathematically as $(x-x_{center})^2 + (y - y_{center})^2 = r^2$ where
$(x_{center},y_{center})$ is the center of the circle, and $r$ is the radius of the circle. From
equation, we can see we have 3 parameters, so we need a 3D accumulator for hough transform, which
would be highly ineffective. So OpenCV uses more trickier method, **Hough Gradient Method** which
uses the gradient information of edges.

The function we use here is **cv.HoughCircles()**. It has plenty of arguments which are well
explained in the documentation. So we directly go to the code.

```python
import cv2 as cv
import ipywidgets as widgets
import numpy as np
from ipycanvas import Canvas, hold_canvas

img = cv.imread("../../../assets/opencv-logo-white.png", cv.IMREAD_GRAYSCALE)

canvas = Canvas(width=img.shape[1], height=img.shape[0])
p1_sl = widgets.IntSlider(value=50, min=1, max=200, description="param1")
p2_sl = widgets.IntSlider(value=30, min=1, max=200, description="param2")
md_sl = widgets.IntSlider(value=20, min=1, max=100, description="minDist")
mr_sl = widgets.IntSlider(value=0, min=0, max=200, description="minRadius")
Mr_sl = widgets.IntSlider(value=200, min=0, max=200, description="maxRadius")


def _update(*_):
    cimg = cv.cvtColor(img, cv.COLOR_GRAY2BGR)
    circles = cv.HoughCircles(
        img,
        cv.HOUGH_GRADIENT,
        1,
        md_sl.value,
        param1=p1_sl.value,
        param2=p2_sl.value,
        minRadius=mr_sl.value,
        maxRadius=Mr_sl.value,
    )
    # Draw the circles detected
    if circles is not None:
        circles = np.uint16(np.around(circles))
        for i in circles[0, :]:
            cv.circle(cimg, (i[0], i[1]), i[2], (0, 255, 0), 2)
            cv.circle(cimg, (i[0], i[1]), 2, (0, 0, 255), 3)
    with hold_canvas():
        canvas.put_image_data(cv.cvtColor(cimg, cv.COLOR_BGR2RGB), 0, 0)


for w in (p1_sl, p2_sl, md_sl, mr_sl, Mr_sl):
    w.observe(_update, "value")
_update()

display(
    widgets.VBox(
        [
            widgets.HBox([p1_sl, p2_sl, md_sl]),
            widgets.HBox([mr_sl, Mr_sl]),
            canvas,
        ]
    )
)
```
